## 전처리

### numeric type으로 변경

 (이거 안해도 데이터 불러오면 float값으로 자동 변경 해주는듯)

```python
for i in range(len(total_df.columns)):
    col = total_df.iloc[0,i]
    print(f'{i+1}th column : "{total_df.columns[i]}"  |  type is {type(col)}')
    if type(col) == str:
        if col.isdigit():
            total_df.iloc[:,i] = total_df.iloc[:,i].astype('int')
            print('changed to int')
        else:
            try:
                total_df.iloc[:,i] = total_df.iloc[:,i].astype('float')
                print('changed to float')
            except:
                print('error')
    else:
        print('not changed')
    print('result:', total_df.iloc[0,i].dtype)
        
```
